```python
from sqlalchemy import create_engine, text

# Substitua as credenciais conforme necessário
host = 'localhost'
port = '5432'
user = 'postgres'
password = 'postgres'
database = 'postgres'  # Use o banco de dados padrão 'postgres' para criar um novo banco de dados

# Criar uma engine SQLAlchemy
engine = create_engine(f'postgresql://{user}:{password}@{host}:{port}')

# Conectar ao PostgreSQL sem iniciar uma transação
with engine.connect() as connection:
    # Verificar se o banco de dados já existe
    result = connection.execute(text(f"SELECT 1 FROM pg_database WHERE datname = '{database}'"))

    # Se o banco de dados já existir, não é necessário criar novamente
    if result.fetchone():
        print(f'O banco de dados "{database}" já existe.')
    else:
        # Executar o comando SQL para criar o banco de dados
        connection.execute(text(f"CREATE DATABASE {database}"))

# Conectar ao banco de dados recém-criado
engine = create_engine(f'postgresql://{user}:{password}@{host}:{port}/{database}')
```
